```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
# ax.spines['left'].set_position('left')
ax.spines['bottom'].set_position('zero')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
```

```python
def shield(d,t,r,x):
    s1 = ((d*t*x) / ((1-t)*r)) 
    s2 = (1-(1+(1-t)*r)**(-1/d))
    return s1 * s2
```

```python
# Illustration of the dependence of the value of the tax shield (vertical axis) 
# on the depreciation rate (horizontal axis). Assumptions: r = 10% and t = 30%.
d = np.linspace(.001,1,1000)
t = .3
r = .1
x = 10000
```

```python
# plot the function
plt.plot(d,shield(d,t,r,x), 'r')
plt.xlabel('Depreciation Rate')
plt.ylabel('Value of Tax Shield')
# show the plot
plt.savefig('f1.pdf')
plt.show()
```
